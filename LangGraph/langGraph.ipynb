# Agentic AI 
is an AI system that can understand a goal, create a plan, take actions, use different tools, remember previous steps, and make decisions to complete a task with minimal human intervention.

Agentic AI wo AI hai jo sirf answer nahi deta, balki ek goal ko achieve karne ke liye khud planning karta hai, decisions leta hai, tools use karta hai aur multiple steps perform karta hai.
